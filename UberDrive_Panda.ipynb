```python
# Find most recent and the earliest travelled date and mean distance travelled for the each city.
```

```python
import pandas as pd
df = pd.read_csv("UberDrive.csv")
```

```python
# Lets remove the last row first 
df = df[:-1]
# Lets convert START_DATE* to Date and time
df["START_DATE*"] = pd.to_datetime(df["START_DATE*"])
df.head(10)

```

```python
res = df.groupby(["START*","STOP*"]).agg({"MILES*" : ["mean"], "START_DATE*" : ["min","max"]})
res.reset_index(inplace = True)
res
res.columns = ["START_CITY", "END_CITY", "AVERAGE_DISTANCE", "EARLIEST_TRIP", "RECENT_TRIP"]
res
```
