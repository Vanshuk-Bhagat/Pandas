```python
# Calculate the sum of sales for each store
import pandas as pd
df = pd.read_csv("store_sales.csv")
df
store_sales_sum = df.iloc[:, 2:].sum(axis=1)  # Skipping 'store_id' and 'city'

# Add the total sales as a new column
df["total_sales"] = store_sales_sum

# Display the updated DataFrame
df[["store_id", "total_sales"]]
```

```python
# Find the store_id of the row with the maximum 'feb' sales

max_feb_sales_store = df.loc[df["feb"].idxmax(), "store_id"]

# Display the result
print(f"Store with the highest sales in February: {max_feb_sales_store}")

```

```python
# Calculate the mean for each column starting from the third column onwards (assuming these are monthly sales data)
mean = df.iloc[:, 2:].mean()
# Display the resulting mean values for each month (jan, feb, etc.)
mean
```

```python
# Count the number of stores in each city
city_store_count = df['city'].value_counts()

# Display the result
print(city_store_count)
```

```python
import pandas as pd

# Step 1: Load the dataset
df = pd.read_csv("store_sales.csv")

# Step 2: Calculate total sales for each store (sum of jan to dec columns)
df["total_sales"] = df.iloc[:, 2:].sum(axis=1)  # Assuming 'jan' to 'dec' are from column 2 onwards

# Step 3: Filter stores where total sales are greater than 500
high_sales_stores = df[df["total_sales"] > 500]

# Step 4: Display the result with store_id and total_sales
print(high_sales_stores[["city", "total_sales"]])
```
